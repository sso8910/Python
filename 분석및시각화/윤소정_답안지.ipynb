#### 문제3 ) 차트를 만드는 코드를 작성하시오

```python
X = [2,3,4,5,6,7,8]
Y1 = [5,7,3,9,1,8,6]
Y2 = [9,5,7,2,3,7,1]

plt.plot(X,Y1, label='Y1', c='r')
plt.plot(X,Y2, label='Y2', c='b')
plt.legend()
plt.title('<실행결과>')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.show()
```

#### 문제4) 주간과 야간에 발생하는 사건에 대해 시각화 하시오.
- 조건 1. 위도는 LAT으로, 경도는 LON으로 표기할 것. 
- 조건 2. x축의 제한은 123부터 133까지로 줄 것. 
- 조건 3. 투명도는 0.1로 줄 것. 
- 조건 4. 주간의 데이터와 야간의 데이터가 구분될 수 있게 할 것.

```python
car=pd.read_csv('Traffic_Accident_2017.csv',encoding='euc-kr')
car
```

```python
car.rename(columns={'경도':'LON','위도':'LAT'},inplace=True)
car
```

```python
car.rename(columns={'경도':'LON','위도':'LAT','주야':'DN'},inplace=True)
car
```

```python
car.loc[car.DN=='주간', 'new']='Day'
car.loc[car.DN=='야간', 'new']='Night'
car
```

```python
# x = [x for x in range(123,134)]
f, ax = plt.subplots(figsize=(8,6))
sns.scatterplot(data=car,x='LON',y='LAT', hue= '주야',alpha=0.1).set(xlim=(123,133))
```
