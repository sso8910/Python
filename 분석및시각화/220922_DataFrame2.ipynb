### 피벗테이블
- 변수명.pivot_table(index = 행이름할변수,columns = 칼럼명할변수, aggfunc ='데이터로 보고 싶은 인자스타일(mean,sum,size등등')

###### 연도별 메달의 수를 피벗테이블로 작성하세요

```python
a = kr.pivot_table(index='Year',columns='Medal', aggfunc='size')
a
```

###### 연도별 메달의 수를 피벗테이블로 작성하세요 -> NAN값은 0으로 채워주세요

```python
a = kr.pivot_table(index='Year',columns='Medal', aggfunc='size').fillna(0)
a
```

###### Sports별로 성별 수를 구해보세요

```python
b = kr.pivot_table(index='Sport',columns='Gender', aggfunc='size')
b
```

#### 피벗테이블 : discipline별로 메달 수를 구해보기

```python
c = kr.pivot_table(index='Discipline',columns='Medal', aggfunc='size')
c
```

```python
df3=df.iloc[:2,:4]
df4= df.iloc[5:,3:]

display(df3)
display(df4)
```

```python
# df3과 df4를 합치기.(단, 둘다 포함되는 연도에 대해서만 고려하기-공통칼럼)
t2=pd.concat([df3,df4],join='inner')
t2
```

```python
# df3과 df4를 합치기.모든 칼럼을 다 포함하고, 결측치가 있으면 0으로 대체하시오.
t2=pd.concat([df3,df4]).fillna(0) #join='outer'가 default값이라 안써도 됨
t2
```
