```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname = 'C:/Windows/Fonts/malgun.ttf').get_name()
rc('font',family=font_name)
```

```python
data1 = pd.read_csv('2014년 졸음운전 교통사고.csv', encoding='euc-kr')
data2 = pd.read_csv('2015년 졸음운전 교통사고.csv', encoding='euc-kr')
data3 = pd.read_csv('2016년 졸음운전 교통사고.csv', encoding='euc-kr')
```

```python
data = pd.concat([data1,data2,data3])
data.set_index('구분',inplace=True)
data
```

```python
#데이터 복붙하기
dt=data.copy() 
dt
```

```python
dt.index = pd.to_datetime(dt.index,format='%Y년%m월')
dt.set_index([dt.index.year,dt.index.month],inplace=True)
dt.index.names=['년도','월']
dt
```

```python
# 연도별 평균 사망자
# 월별 평균 사망자
# 힌트 : groupby()
y = dt.groupby('년도').mean()[['사망(명)']]
y
```

```python
m = dt.groupby('월').mean()[['사망(명)']]
m
```

```python
#2014년도 월별 사망, 부상 데이터를 바 차트로 그려주세요.
dt.loc[2014][['사망(명)','부상(명)']].plot(kind='bar',rot=0)
plt.show()
```
