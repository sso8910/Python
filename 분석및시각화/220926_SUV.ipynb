```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname = 'C:/Windows/Fonts/malgun.ttf').get_name()
rc('font',family=font_name)
```

```python
suv=pd.read_csv('suv.csv',encoding='utf-8')
suv
```

```python
suv.info()
```

```python
#get_dummies 함수를 써서 'Gender변수'를 0,1값으로 부여함(결측값을 제외하고 0,1로 구성된 더미값을 만들어 줌)
#dummy_na=True를 사용하면 결측값도 인코딩해서 처리해줌
suv2=pd.get_dummies(suv,columns=['Gender'],dummy_na=True)
suv2
```

```python
#age변수 분포 확인(정규분포)
f, ax = plt.subplots(figsize=(10,6))
sns.distplot(suv2['Age'])
```
