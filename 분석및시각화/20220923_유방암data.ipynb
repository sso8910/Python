```python
#문자열을 숫자로 변환해주려면? m=1, b=0
#변환해준 열을 '판정'으로 이름 짓고 맨 뒤에 칼럼추가하기
#df1에 판정이라는 칼럼을 추가할꺼야~ 
def func1(row) :
    if 'M' in row:
        return 1
    else :
        return 0
df1['판정']=df1['diagnosis'].apply(func1)
```
