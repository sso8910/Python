### matplotlib(시각화)라이브러리
- 광범위하게 분산된 방대한 양의 자료를 한 눈에 볼 수 있도록 도표나 차트로 정리
- 시각화를 통해 데이터의 특징을 쉽게 파악 가능
- 분석 결과를 상대방에게 효과적으로 전달 가능
- pyplot :시각화 , 비대화형(간단한 정보만 입력)
- pylab : 시각화+numpy
- import matplotlib.pyplot as plt : matplotlib모듈에서 pyplot을 import하고 plt라는 이름으로 부르겠다.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

#### Line Plot 그리기

```python
y = [2,4,6]  #X 값은 지정안했으니까, 랜덤으로 지정
plt.plot(y)
plt.show()
```

#### Line Style 설정하기

```python
x = [1,2,3]
y = [2,4,6]  
plt.plot(x,y, ls='--')   
plt.show()
#'-','--','-.',':' 종류가 있다
```

#### Marker 찍기(marker='@' , Line Width(라인넓이 lw=원하는 숫자)
- line color 변경(color = 'r')
- marker color변경(markerfacecolor='@')
- 축변경 plt.xlim(구간,구간), plt.ylim(구간,구간)
- plt.grid() 격자 추가

```python
x = [1,2,3]
y = [2,4,6]  
plt.plot(x,y, ls='--', marker='o', lw = 2,
        color = 'b', mec='b', mew=3, mfc='w')   
plt.show()
```

### 연습문제 : 요일별 교통사고 시각화

```python
data = pd.read_csv('Traffic_Accident_2017.csv', encoding='euc-kr')
data
```

```python
#요일별 사고 건수를 세봅시다
temp = data['요일'].value_counts()
temp
```

```python
y = temp[['월','화','수','목','금','토','일']]
y
```

```python
x = y.index
plt.bar(x,y)
plt.title('2017년 요일별 사망교통사고 건수')
plt.xlabel('요일')
plt.ylabel('사고건수').set_rotation(0)
plt.ylim(500,620)
plt.show()
```

### 발생지시도 별 사고건수 그래프 그리기

#### '사고유형_대분류'칼럼에서 차대차만 뽑아오기

```python
car = data[data['사고유형_대분류']=='차대차']
car.head(3)
```

```python
#car에서 '발생지시도'를 기준으로 '사상자수'의 합계 구하기
place = car[['사상자수','발생지시도']].groupby('발생지시도').sum()
place
```

```python
x = place.index
y = place.values
plt.plot(x,y)
plt.xlabel('지역')
plt.ylabel('사상자수').set_rotation(0)
plt.title('차 vs 차 교통사고 사상자 수')
plt.show()
```

#### 파이차트
- 교통사고가 가장 많이 발생하는 시간대를 알아봅시다.
- 0~2, 3~5, 6~8, ....., 21~23시

```python
time=car['발생년월일시']
time%=100
time
```

```python
#구간정하기
bins = [-1,2,5,8,11,14,17,20,23]
labels = ['0~2','3~5','6~8','9~11','12~14','15~17','18~20','21~23']
result = pd.cut(time,bins,labels=labels)
result
```

```python
rs = result.value_counts().sort_index()
rs
```

```python
plt.figure(figsize=(5,5))
plt.pie(rs,labels=labels,autopct="%1.1f%%",counterclock=False)
plt.show()
#counterclock=False 넣으면 내용의 순서(위치)가 바뀜 
```

#### 파이차트 포인트만 분리시키고 음영주기
- explode = []....] 중심에서 얼마나 떨어질지 적기
- shadow : 음영주기

```python
rs = result.value_counts().sort_index()
plt.figure(figsize=(5,5))
explode = [0,0.10,0,0.10,0,0.10,0,0.10]
plt.pie(rs,labels=labels,autopct="%1.1f%%",explode=explode,shadow=True)
plt.show()
```
