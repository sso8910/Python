```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
from bs4 import BeautifulSoup as bs
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
import time
from tqdm import tqdm_notebook as tqdm
import requests as req
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get('https://en.wikipedia.org/wiki/List_of_coffee_varieties')
time.sleep(3)
```

```python
name_list=[]

for i in tqdm(range(1,45)):
    name = driver.find_elements_by_css_selector('#mw-content-text > div.mw-parser-output > table.wikitable.sortable.jquery-tablesorter > tbody > tr:nth-child({}) > td:nth-child(1)'.format(i))
    for i in name:
        name_list.append(i.text)
    time.sleep(1)
```

```python
species_list= []

for i in tqdm(range(1,45)):
    species = driver.find_elements_by_css_selector('#mw-content-text > div.mw-parser-output > table.wikitable.sortable.jquery-tablesorter > tbody > tr:nth-child({}) > td:nth-child(2)'.format(i))
    for i in species:
        species_list.append(i.text)
    time.sleep(1)
```

```python
reigion_list =[]

for i in tqdm(range(1,45)):
    reigion = driver.find_elements_by_css_selector('#mw-content-text > div.mw-parser-output > table.wikitable.sortable.jquery-tablesorter > tbody > tr:nth-child({}) > td:nth-child(3)'.format(i))
    for i in reigion:
        reigion_list.append(i.text)
    time.sleep(1)
```

```python
common_list= []
for i in tqdm(range(1,45)):
    common = driver.find_elements_by_css_selector('#mw-content-text > div.mw-parser-output > table.wikitable.sortable.jquery-tablesorter > tbody > tr:nth-child({}) > td:nth-child(4)'.format(i))
    for i in common:
        common_list.append(i.text)
    time.sleep(1)
```

```python
nltk.corpus.stopwords.words('english')
stopwords = nltk.corpus.stopwords.words('english')
```

```python
stopwords = ['i', 'be','see','lol.i','see','beans.i',
            'have','want','joe','look','bit','continue','winner.this',
            '”','t','bulletproof','say','“','do','me.i','[',']']
```

```python
def extract_tags_english(string):
    word_list = []
    string = string.lower()
    string = nltk.word_tokenize(string)
    string = nltk.pos_tag(string)
       
    for word, tag in string:
        if tag in ['NN', 'VB']:
            if word not in stopwords: #if word not in stopwords:
                word_list.append(word)

    return word_list
```

```python
common_list_nltk_update = extract_tags_english(common_list_nltk)
common_list_nltk_update
```

```python
count = Counter(common_list_nltk_update)
count
```
