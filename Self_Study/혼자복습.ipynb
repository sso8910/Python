### requests
- 크롤링하기 전 페이지의 권한을 확인하는 작업.
- res 변수에 작업하려는 화면을 담고, 200이 나오면 정상코드. 
- requests코드 400번은 에러나기 때문에 긁어올 수 없음.

```python
import requests as req #제일 먼저 실행해야함
```

```python
res = req.get('https://www.google.com/')
res.raise_for_status()  #사이트 상태가 정상이면 크롤링하고, 아니면 에러창나옴
res
```
