###### 00중학교 1학년 1반의 중간고사 평균 성적(국어, 영어, 수학, 과학, 사회) 점수를 Series로 생성해 주세요.
- 국어 75점, 영어 80점, 수학 90점, 과학 95점, 사회 70점
- Series로 생성 -> pd.Series([변수값,,,,,],index=['변수명....])

```python
midTerm_score = pd.Series([75,80,90,95,70],index=['국어','영어','수학','과학','사회'])
midTerm_score
```

```python
midTerm_score.index.name = '과목'
midTerm_score.name = '중간고사 성적'
midTerm_score
```

```python
print(midTerm_score[0:3])
print(midTerm_score['국어':'수학'])
```

#### 중간고사 평균 성적이 85점 이상인 과목만 뽑아주세요

```python
midTerm_score[(midTerm_score>=85)]
```

#### 중간고사 평균 성적이 80점 이상 90점 이하인 과목만 뽑아주세요

```python
midTerm_score[(midTerm_score>=80)&(midTerm_score<=90)]
```

#### 5개 과목의 평균, 합계, 최대값 출력해주세요. numpy import 해주기

```python
print(np.mean(midTerm_score))
print(np.sum(midTerm_score))
print(np.max(midTerm_score))
```

```python
#quantity,item_price 총 2개의 컬럼으로 구성된 데이터 프레임을 만들어주세요
a=df[['quantity','item_price']]
a
```

```python
# item_price의 $표시를 없애주고 float 타입으로 바꾸고 그 결과를 'new_price'라는 컬럼에 저장해주세요.
df['new_price']=df['item_price'].str[1:].astype('float64')
a=df['new_price'].head()
```

```python
a
```

```python
#new_price 칼럼이 5이하인 값을 가지는 데이터 프레임을 추출하고 그 갯수를 구해주세요
b=len(df.loc[df.new_price<=5])
b
```

```python
#item_name이 Chicken Salad Bowl인 데이터 프레임을 추출하고, index값을 초기화해주세요.
#drop=true 하면 기존의 index는 삭제하고 새로 index를 부여해줌
c=df.loc[df.item_name =='Chicken Salad Bowl'].reset_index(drop=True)
c.head(2)
```

```python
#item_name이 Chicken Salad Bowl'이고, new_price값이 10이하인 데이터 프레임을 추출
d=df.loc[(df.item_name =='Chicken Salad Bowl') & (df.new_price<=10)].reset_index(drop=True)
d.head(3)
```

```python
#new_price 컬럼의 값을 오름차순으로 정리하고, index를 초기화해주세요.
e=df.sort_values('new_price').reset_index(drop=True)
e
```

```python
#df의 item_name에서 메뉴 이름에 'Chips'가 들어가있는 데이터 출력
f=df.loc[df.item_name.str.contains('Chips')]
f.head(6)
```

```python
#df의 new_price 컬럼 값들을 내림차순으로 정리하고, index를 초기화해주세요
g=df.sort_values('new_price')[::-1].reset_index(drop=True)
#g=df.sort_values('new_price',ascending=False).reset_index(drop=True)
g.head(3)
```

```python
#df의 item_name 컬럼 값이 Steak Salad 또는 Bowl인 데이터를 뽑아주세요
h = df.loc[df.item_name.str.contains('Steak Salad')|df.item_name.str.contains('Bowl')]
#h = df.loc[(df.item_name=='Steak Salad')|(df.item_name=='Bowl')]
h
```

```python
#중복제거
i = df.loc[df.item_name.str.contains('Steak Salad')|df.item_name.str.contains('Bowl')]
i = i.drop_duplicates('item_name', keep = False)
i
#keep=False : 중복되는 애는 다 삭제시킬거야
```
