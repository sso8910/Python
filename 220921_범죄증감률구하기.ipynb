### 범죄 증감률 계산하기

```python
import pandas as pd
import pandas as np
```

```python
df2015=pd.read_csv('2015.csv',encoding='cp949',index_col='관서명')
df2016=pd.read_csv('2016.csv',encoding='cp949',index_col='관서명')
df2017=pd.read_csv('2017.csv',encoding='cp949',index_col='관서명')
```

```python
df2017
```

#### 2017년도에만 있는 데이터 삭제

```python
df2017=df2017.drop(['광주지방경찰청'])
```

```python
# 구분이 발생건수인 값들만 가져오기(구분==발생건수)
# 구분칼럼의 값이 발생건수인 데이터만 가져오기
df2015_crime = df2015[df2015['구분']=='발생건수']
df2015_crime['2015총계']=df2015_crime.sum(axis=1)
df2015_crime
```

```python
df2016_crime = df2016[df2016['구분']=='발생건수']
df2016_crime['2016총계']=df2016_crime.sum(axis=1)
df2016_crime
```

```python
df2017_crime = df2017[df2017['구분']=='발생건수']
df2017_crime['2017총계']=df2017_crime.sum(axis=1)
df2017_crime
```

#### 데이터 총합 열 생성

```python
s1=df2015_crime['2015총계']
s3=df2016_crime['2016총계']
s5=df2017_crime['2017총계']
```

```python
s3
```

#### 전년대비 증감율 계산하고 이름변경해주기

```python
s2=(s3-s1)/s1*100
s2.name='2015-2016 증감율'
#Series의 name값은 DataFrame에서 Column값으로 사용됨
s4=(s5-s3)/s3*100
s4.name='2016-2017 증감율'
```

```python
s2
```

#### 행 기준으로 데이터 합치기

```python
pd.concat([s1,s2,s3,s4,s5],axis=1)
```

#### Concat vs Merge
- 3개의 Dictionary 생성

```python
employee = [{'empno':1, 'ename':'kim', 'dept':1},
            {'empno':2, 'ename':'lee', 'dept':2},
            {'empno':3, 'ename':'park', 'dept':1},
            {'empno':4, 'ename':'song', 'dept':3},
            {'empno':5, 'ename':'min', 'dept':2}
           ]
dept=[{'dept':1, 'deptname':'관리직'},
      {'dept':2, 'deptname':'영업직'},
      {'dept':3, 'deptname':'개발직'}
     ]
info =[{'empno':1, 'addr':'서울시','phone':'010-1111-1111'},
       {'empno':3, 'addr':'부산시','phone':'010-2222-2222'},
       {'empno':2, 'addr':'광주시','phone':'010-3333-3333'},
       {'empno':5, 'addr':'광주시','phone':'010-4444-4444'},
       {'empno':4, 'addr':'광주시','phone':'010-5555-5555'}
      ]
```

```python
emp = pd.DataFrame(employee)
dept = pd.DataFrame(dept)
info = pd.DataFrame(info)
```

#### merge : 공통 칼럼이 존재해야하며, 그 공통칼럼을 기준으로 데이터프래임을 병합시킴
- 병합시키면 자동으로 정렬되고, 공통칼럼이 2개 이상인 경우 정렬 기준을 정해주는건 on='변수명'
- 병합시키고 싶을 때, 공통 칼럼을 알면 merge, 모르면 concet 쓰는게 좋음

```python
pd.merge(emp,dept)
```

```python
pd.merge(emp,dept,on='dept')
```

```python
pd.merge(dept,emp,on='dept')
```

```python
mg = pd.merge(emp,dept,on='dept')
mg.drop(columns=['dept'])
mg
#삭제한 상태로 보이고 껏다키면 다시 살아남
```

#### inlpace=True : 데이터 프레임이 바뀐 상태로 저장됨, 
- 삭제하고 싶은 데이터를 삭제한게 저장되서 유지됨

```python
mg = pd.merge(emp,dept,on='dept')
mg.drop(columns=['dept'],inplace=True)
mg
```

```python
mg1= pd.merge(emp,dept,on='dept')
mg1.drop(columns=['dept'],inplace=True)
mg1
```
